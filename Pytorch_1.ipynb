###Batch dimensions and Broadcasting
Rules differ as we multiply tensors with dimensions higher than 2.
The rules can be simplified as:
- Look at the last to dimensions of the tensor. Everything before them is treated as batch dimensions to be broadcasted.
- If the last two dimensions of the matrices follow the basic matrix multiplication of rule, the first check is completed.
- Then, we need to look at the batch dimensions. Check from right to left->If corresponding dimensions are equal or one of them is one then they can be broadcasted. The second check is complete.
- If the tensors follow the two checks then they can be multiplied.

**Note:** If there is a missing batch dimensions i.e., dimensions dont match then pytorch conceptually pads the missing batch dimensions with 1 to the left.

For example: If A(2,3,4) and B(4, 3) then we assume B as (1, 4, 3).

**Broadcasting**- It the pytorch's way of saying that it will automatically expand the smaller tensor if the dimensions follow the rules( 1 or equal).

```python
multensor3 = torch.rand(10, 20, 20, 30)
multensor4 = torch.rand(1, 20, 30, 50 )
result2 = torch.matmul(multensor3, multensor4)
result2.ndim, result2.shape
```

###RESHAPING, STACKING, SQUEEZING, UNSQUEEZING, AND PERMUTING TENSORS
As we have already mentioned several times, shape errors are among the most common errors we encounter in DL. To avoid this, we can reshape the tensor according to our need.

1. Reshape - Use torch.reshape(input, shape) to reshape input to shape. The shape must be valid according to the original shape of the tensor.
2. View - Use tensor.view(shape) to view the tensor in specified shape. View shares the same memory space as the original tensor, so changing the view changes the tensor.
3. Stack - Use torch.stack(tensors, dim) to stack tensors (concatenate) at the specified dimension
4. Squeeze - Use torch.squeeze(input) to remove all dimensions with value 1.
5. Unsqueeze - Use torch.unsqueeze(input, dim) to add 1 dimension at the specified dimension.
6. Permute - Returns a view of the original tensor with dimensions permuted in a certain order.

```python
original_tensor = torch.rand(1, 3, 4)
original_tensor
```

```python
reshaped_tensor = original_tensor.reshape(6, 2)
reshaped_tensor
```

```python
#Creating a view with another shape
view_tensor = reshaped_tensor.view(3, 4)
view_tensor
```

```python
#Stacking tensors upon tensors
stack_tensor = torch.rand(3, 4)
stacked_tensor = torch.stack([view_tensor, stack_tensor], dim = 1)
stacked_tensor
```

```python
image_tensor_permute = torch.rand(3, 244, 244)
permuted = image_tensor_permute.permute(1, 2, 0)
permuted.shape
```

###REPRODUCIBILITY
Neural networks usually start with randomly initialized tensors (such as weights and biases). Since these tensors are random, training the same model multiple times can produce different results.

To make the random numbers reproducible, PyTorch provides the function:

- torch.manual_seed(seed)

          where seed is any integer.

Setting the same seed causes PyTorch's random number generator to produce the same sequence of random numbers every time the program is run. This means tensors created after setting the seed will have the same values across runs (provided the rest of the code is the same).
torch.manual_seed() does not remove randomness. It simply makes the randomness deterministic (repeatable).

You do not need to call torch.manual_seed() before every tensor.

- For example:

    import torch

    torch.manual_seed(42)

    A = torch.rand(2, 2)
    B = torch.rand(2, 2)
    C = torch.rand(2, 2)

Here, A, B, and C will always have the same values every time you run the program because they are generated from the same deterministic sequence.

You only need to call torch.manual_seed() again if you want to restart the random number sequence.

torch.manual_seed(42)
A = torch.rand(2, 2)

torch.manual_seed(42)
B = torch.rand(2, 2)

print(torch.equal(A, B))  # True

```python
RANDOM_SEED = 45
torch.manual_seed(RANDOM_SEED)
seed_tensor = torch.rand(3, 4)

torch.manual_seed(RANDOM_SEED)
seed_tensor2 = torch.rand(3, 4)

print(seed_tensor)
print(seed_tensor2)
print(seed_tensor == seed_tensor2)
```
