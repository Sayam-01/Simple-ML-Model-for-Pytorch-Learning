#Computer Vision
Computer Vision is one the most sought after paradigm of ML. Basically, it is inferring information from visual data like images, videos etc.

Convulational Neural Networks(CNNs) were the mose domination and efficient ML models for computer vision. They have been replaced - in some cases - by Visual Transformers.

##CNNs
It is a specific neural network designed to process grid like data- especially images. The reason we don't use normal neural networks is because it would take too much memory,overfitting, and their inability to learn image structures.

Suppose you have a 224 × 224 RGB image.

- Height = 224 pixels
- Width = 224 pixels
- Channels = 3 (Red, Green, Blue)

The total number of input values is:

- 224×224×3=150,528

If you connect these inputs to just 1,000 neurons in the first hidden layer:

- 150,528×1,000=150,528,000

That's over 150 million weights in only the first layer. Simply, too much.

### - How CNN works:
Instead of looking at the entire image at once, CNNs will look at small regions at once using filter known as kernel.

Take this filter as an example.

The filter will slide across the regions.

At each region it will calculate the weighted sums of the pixels, which is known as convulation. The result of these operations will be a feature map.
